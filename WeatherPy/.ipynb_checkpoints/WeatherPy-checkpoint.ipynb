## 1.) Dependencies and Setup

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from random import sample

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"
output_png_path = "output_data/plots/"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## 2.) Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
# Select a random sample of 500 cities from list generated in above cell
cities_sample = sample(cities,500)
```

## 3.) Generate API Calls to Create Dataframe

```python
# Create API calls to pull current weather data
base_url = "http://api.openweathermap.org/data/2.5/weather?q="
units = 'imperial'

city_name = 'busselton'


query_url = f'{base_url}{city_name}&appid={weather_api_key}&units={units}'

response = requests.get(query_url).json()
```

```python
# Loop through list of cities in sets of 50, append records to dataframe
# Sleep for 30 seconds in between each set

# Create blank dataframe with relevant fields, so that records from loop can be added in
column_names = ['City','Lat','Lng','Max Temp','Humidity','Cloudiness','Wind Speed','Country','Date']
city_weather_df = pd.DataFrame(columns=column_names)

# ---------- BEGIN LOOP FOR BULK API CALL ----------
# Set starting index # for city loop, so that you continue through each city list as you progres to subsequent set
city_loop_index = 0
for loop_set in range(0,10):
    print(f'--------- Set {loop_set + 1} of 10 ---------- ')
    # Loop through cities and make API calls
    for city_index in range(0,50):
        # Add city_loop_index variable to ensure that loop continues through list of 500
        final_index = city_index + city_loop_index
        # Pull current city
        city_name = cities_sample[final_index]
        # Generate query string
        query_url = f'{base_url}{city_name}&appid={weather_api_key}&units={units}'
        # Make API Call
        response = requests.get(query_url).json()
        # Create try/except block to skip record if city is not found
        try:
            # Extract relevant data from response
            latitude = response['coord']['lat']
            longitude = response['coord']['lon']
            max_temp = response['main']['temp_max']
            humidity = response['main']['humidity']
            cloudiness = response['clouds']['all']
            wind_speed = response['wind']['speed']
            country = response['sys']['country']
            date = response['dt']
            # Create dictionary to append to df
            record_dict = {
                'City':city_name,
                'Lat':latitude,
                'Lng':longitude,
                'Max Temp':max_temp,
                'Humidity':humidity,
                'Cloudiness':cloudiness,
                'Wind Speed':wind_speed,
                'Country':country,
                'Date':date
            }
            # Append record dict to city_weather_df
            city_weather_df = city_weather_df.append(record_dict,ignore_index=True)
            print(f'City {city_name} found successfully as record {city_index + 1} of set {loop_set + 1}.')
        except (KeyError,IndexError):
            print(f'City {city_name} not found. Skipping...')
    # ---------- Outer Loop ----------
    # Add 50 to city_loop_index to ensure that inner loop continues with next set of 50 cities
    city_loop_index += 50
    # Sleep for 30 seconds in between each set
    time.sleep(30)

print('---------- Data Retrieval Complete ----------')
            
```

## 5.) Generate Scatter Plots and Linear Regression - Northern vs. Southern Hemisphere

```python
# Separate out dataframe into northern hemisphere and southern hemisphere
northern_hemisphere_df = city_weather_df.loc[city_weather_df['Lat'] >= 0,:]
southern_hemisphere_df = city_weather_df.loc[city_weather_df['Lat'] < 0,:]
```

**Northern Hemisphere: Temperature (F) vs. Latitude Linear Regression**

```python
# Create scatter plot

northern_temp_vs_latitude_scatter = northern_hemisphere_df.plot.scatter(x='Lat',y='Max Temp')
northern_temp_vs_latitude_scatter.set_xlabel("Latitude")
northern_temp_vs_latitude_scatter.set_ylabel("Temperature (F)")
northern_temp_vs_latitude_scatter.set_title ("Temperature (F) vs Latitude in Northern Hemisphere (4/22/2021)")

plt.show()

```

```python
# Calculate linear regression model

# Create series for independent and dependent variables
latitude = northern_hemisphere_df['Lat']
max_temp = northern_hemisphere_df['Max Temp']

# Create linear regression model
calc_slope, calc_int, calc_r, calc_p, calc_std_err = linregress(latitude, max_temp)
line_best_fit = calc_slope * latitude + calc_int

# Add line of best fit to above plot
plt.plot(latitude,line_best_fit,'r--')

# Add regression equation and r^2 value to above plot
line_eq = "y = " + str(round(calc_slope,2)) + "x + " + str(round(calc_int,2))
plt.annotate(line_eq,(0,25),fontsize=15,color="red")
r_squared = calc_r ** 2
plt.annotate('r-squared: ' + str(round(r_squared,4)),(0,15),fontsize=15,color="red")

# Add analysis to figure
# NOTE: approach taken from https://stackoverflow.com/questions/43158914/issues-with-adding-figtext-to-bottom-of-subplots
text = '\n\n\n\nIn the northern hemisphere, as latitude (independent variable) increases, the temperature (dependent variable) decreases. Thus, the regression model produced a negative slope. The r-squared value is 0.78, indicating a fairly strong fit.'
plt.subplots_adjust(bottom=0.3)
plt.figtext(x=.05,y=.1,s=text,wrap=True,fontsize=7)

plt.savefig(output_png_path + 'northern_hemisphere_temp_vs_latitude_regression.png')
```

**Southern Hemisphere: Temperature (F) vs. Latitude Linear Regression**

```python
# Create scatter plot

southern_temp_vs_latitude_scatter = southern_hemisphere_df.plot.scatter(x='Lat',y='Max Temp')
southern_temp_vs_latitude_scatter.set_xlabel("Latitude")
southern_temp_vs_latitude_scatter.set_ylabel("Temperature (F)")
southern_temp_vs_latitude_scatter.set_title ("Temperature (F) vs Latitude in Sothern Hemisphere (4/22/2021)")

plt.show()
```

```python
# Calculate linear regression model

# Create series for independent and dependent variables
latitude = southern_hemisphere_df['Lat']
max_temp = southern_hemisphere_df['Max Temp']

# Create linear regression model
calc_slope, calc_int, calc_r, calc_p, calc_std_err = linregress(latitude, max_temp)
line_best_fit = calc_slope * latitude + calc_int

# Add line of best fit to above plot
plt.plot(latitude,line_best_fit,'r--')

# Add regression equation and r^2 value to above plot
line_eq = "y = " + str(round(calc_slope,2)) + "x + " + str(round(calc_int,2))
plt.annotate(line_eq,(-55,82),fontsize=15,color="red")
r_squared = calc_r ** 2
plt.annotate('r-squared: ' + str(round(r_squared,4)),(-55,77),fontsize=15,color="red")

# Add analysis to figure
# NOTE: approach taken from https://stackoverflow.com/questions/43158914/issues-with-adding-figtext-to-bottom-of-subplots
text = '\n\n\n\nIn the southern hemisphere, as latitude (independent variable) increases, the temperature (dependent variable) increases. Thus, the regression model produced a positive slope. The r-squared value is 0.53, meaning that 53% of the variation in temperature in the southern hemisphere can be explained by the variation in latitude.'
plt.subplots_adjust(bottom=0.3)
plt.figtext(x=.05,y=.1,s=text,wrap=True,fontsize=7)

plt.savefig(output_png_path + 'southern_hemisphere_temp_vs_latitude_regression.png')
```
